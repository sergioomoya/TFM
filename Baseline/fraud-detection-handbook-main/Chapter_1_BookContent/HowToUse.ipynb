# Cómo usar este libro


Este libro es un [Jupyter Book](https://jupyterbook.org/intro.html), que le permite ejecutar, reproducir o modificar interactivamente las secciones de este libro que contienen código. Este capítulo proporciona una descripción general de qué son los libros de Jupyter y cómo permiten que su contenido sea reproducible. 


## Proyecto Jupyter Book

El proyecto Jupyter Book es un proyecto de código abierto. Es desarrollado por [Executable Book Project](https://executablebooks.org/en/latest/), que es una [colaboración internacional entre varias universidades y otros proyectos de código abierto](https://executablebooks.org/en/latest/about.html). 

Debajo del capó, un libro de Jupyter es esencialmente una colección de archivos markdown y cuadernos de Jupyter. Estos archivos se compilan juntos de acuerdo con la configuración y las especificaciones de la tabla de contenido. Puede seguir el icono de Github en la parte superior de esta página para ver el código fuente de este libro. 


## Diseño del libro

El diseño del libro proporciona cinco zonas principales:

1. La herramienta de búsqueda (arriba a la izquierda): Permite encontrar todas las secciones del libro que contienen un término en particular.
2. La tabla de contenido (panel izquierdo): Permite navegar entre los diferentes capítulos y secciones principales del libro.
3. Los enlaces externos (arriba): Contiene iconos para iniciar cuadernos, cambiar al modo de pantalla completa, acceder al repositorio de Github o descargar el contenido de la sección actual.
4. El panel principal (centro): El contenido de la sección actual.
5. El contenido de la sección (panel derecho): Permite navegar en las subsecciones de la sección actual. 




## Reproducibilidad

Todas las secciones que contienen código (cuadernos Jupyter), como esta sección, tendrán un icono de transbordador en los enlaces externos, lo que permite ejecutar el cuaderno en la nube. Se proporcionan dos servicios: Google Colab y Binder. Describimos brevemente estos dos servicios a continuación, con sus pros y contras. 

### Google Colab


[Google Colab](https://colab.research.google.com/notebooks/intro.ipynb#) es un servicio proporcionado por Google Research y requiere tener una cuenta de Google. Permite ejecutar cuadernos de Python de forma gratuita en Google Cloud. 

Los recursos computacionales proporcionados por Google Colab pueden fluctuar. Dependen de los recursos disponibles de Google, así como de su uso del servicio (el uso intenso puede dar lugar a menos recursos). Sin embargo, puede esperar tener un núcleo y algunos gigabytes de RAM. Google Colab también proporciona una GPU gratuita, cuyo modelo puede variar según los recursos disponibles. Los modelos de GPU proporcionados a menudo incluyen Nvidia K80, T4, P4 o P100. Consulte este [enlace](https://research.google.com/colaboratory/faq.html?hl=ko-KRUbuntu) para obtener más detalles sobre los recursos que Google puede proporcionar. 

### Binder

[Binder](https://mybinder.readthedocs.io/en/latest/index.html) {cite}`beg2021using` permite crear, utilizar y compartir entornos informáticos personalizados. Está impulsado por [BinderHub](https://github.com/jupyterhub/binderhub), que es una herramienta de código abierto que implementa el servicio Binder en la nube. 

Los cuadernos se pueden iniciar y ejecutar de forma gratuita, gracias a la [Federación BinderHub](https://mybinder.readthedocs.io/en/latest/about/federation.html), actualmente respaldada por Google Cloud Platform, OVH, cuadernos GESIS y el Instituto Alan Turing. BinderHub se basa en [Repo2Docker](https://github.com/jupyterhub/repo2docker) para generar imágenes de docker desde un repositorio Git, y [JupyterHub](https://github.com/jupyterhub/jupyterhub) para ejecutar y compartir cuadernos. 

Por lo general, Binder tarda uno o dos minutos en crear un contenedor Docker y comenzar a servir un cuaderno. El archivo `environment.yml`, presente en la raíz del repositorio de libros de Github, se utiliza para especificar qué bibliotecas de Python deben instalarse en el contenedor docker. 

Al igual que con Google Colab, los recursos proporcionados de forma gratuita por la Federación BinderHub fluctúan. Consulte la [guía del usuario de Binder](https://mybinder.readthedocs.io/en/latest/about/about.html) y las [pautas para el usuario de Binder](https://mybinder.readthedocs.io/en/latest/about/user-guidelines.html) para obtener más detalles sobre los recursos que la Federación BinderHub puede proporcionar. 


### Pros y contras

A continuación se proporciona un resumen de las características de Google Colab y Binder. 


|| Google Colab | Binder|
:----- | :---- | :-----
|Registro   | Requiere inicio de sesión de Google | Ninguno|
|Recursos de CPU | Fluctúan  | Fluctúan|
|Recursos de RAM | Fluctúan  | Fluctúan|
|Recursos de GPU | Fluctúan. A menudo incluye Nvidia K80s, T4s, P4s o P100s  | Ninguno|
|Duración de la sesión  | Hasta 12 horas  | Hasta 12 horas|
|Entono  | Python 3.6 | Personalizable con archivo environment.yml|

En nuestra experiencia, los tiempos de ejecución de Google Colab son aproximadamente el doble de rápidos que los de Binder. 

Este resumen refleja el estado de los dos servicios en el momento de escribir esta sección (enero de 2021). Sin embargo, se espera que las características de estos dos servicios evolucionen en un futuro próximo.

## Ejecución local y compilación del libro

Este es un libro de código abierto y el código está [disponible en GitHub](https://github.com/Fraud-Detection-Handbook/fraud-detection-handbook) (o siguiendo el icono de GitHub en los enlaces externos en la parte superior de esta página). Puede ejecutar los cuadernos del libro en su ordenador clonando el repositorio de GitHub. 

También puede recompilar el libro. Compilar el libro primero requerirá que [instale Jupyter Book](https://jupyterbook.org/intro.html#install-jupyter-book).

Una vez hecho esto, es un proceso de dos pasos:

1. Clonar el repositorio del libro:

```
git clone https://github.com/Fraud-Detection-Handbook/fraud-detection-handbook
```

2. Compilar el libro

```
jupyter-book build fraud-detection-handbook
```

El libro estará disponible localmente en `fraud-detection-handbook/_build/html/index.html`.
